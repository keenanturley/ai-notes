### Binomial distribution

Probability mass function
$$
\binom{n}{k}p^k(1-p)^{n-k}
$$

Binomial coefficient
$$
\binom{n}{k} = \dfrac{n!}{k!(n-k)!}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def pmf(k, n, p):
    # n! / k!(n-k)!
    binomial_coefficient = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    # probability mass funcion
    return binomial_coefficient * (p ** k) * (1 - p) ** (n-k)
     
red = np.array([pmf(k, 40, 0.5) for k in range(41)])
green = np.array([pmf(k, 20, 0.7) for k in range(21)])
blue = np.array([pmf(k, 20, 0.5) for k in range(21)])

fig, ax = plt.subplots()
fig.set_dpi(200)
ax.set_ylim(-0.01, 0.25)
ax.plot(red, 'ro', markersize=3, label='p=0.5 and n=40')
ax.plot(green, 'gs', markersize=3, label='p=0.7 and n=20')
ax.plot(blue, 'bD', markersize=3, label='p=0.5 and n=20')
fig.legend(bbox_to_anchor=(0.9, 0.875), bbox_transform=fig.transFigure)
plt.show()
```
