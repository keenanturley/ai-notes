# Log-log plot
[Wikipedia](https://en.wikipedia.org/wiki/Log%E2%80%93log_plot)

A two-dimensional graph of numerical data that uses logarithmic scales on both the horizontal and vertical axes. Power functions - relationships of the form y = ax^k - appear as straight lines in a log-log graph, with the exponent corresponding to the slope, and the coefficient corresponding to the intercept.

This notebook recreates the below graph in matplotlib

![](https://upload.wikimedia.org/wikipedia/commons/9/92/LogLog_exponentials.svg)

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
import math

def plot_functions(functions):
    x = np.geomspace(0.1, 100)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))
    fig.set_dpi(200)

    ax1.set_title("Normal Plot")
    ax1.grid(True)
    ax1.set_xlim(0, 10)
    ax1.set_ylim(0, 10)
    [ax1.plot(x, [y["f"](x_val) for x_val in x], label=y["name"]) for y in functions]
    ax1.legend()

    ax2.set_title("Log-log Plot")
    ax2.grid(True)
    ax2.set_xscale("log")
    ax2.set_xlim(0.1, 100)
    ax2.set_yscale("log")
    ax2.set_ylim(0.1, 100)
    [ax2.plot(x, [y["f"](x_val) for x_val in x], label=y["name"]) for y in functions]
    ax2.legend()

    # Customize the tick labels
    def log_tick_formatter(val, pos):
        return f"{val:.1f}"

    ax2.xaxis.set_major_formatter(FuncFormatter(log_tick_formatter))
    ax2.yaxis.set_major_formatter(FuncFormatter(log_tick_formatter))

    plt.show()

functions = [
    {
        "name": "f(x) = x",
        "f": lambda x: x,
    },
    {
        "name": "f(x) = x^2",
        "f": lambda x: x**2,
    },
    {
        "name": "f(x) = x^3",
        "f": lambda x: x**3,
    },
]

plot_functions(functions)
```
